## Prediction Using Supervised Machine Learning

###  Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

### Reading the Dataset

```python
df = pd.read_csv("http://bit.ly/w-data")
df.head()
```

```python
df.info()
```

### Splitting the dataset into the Training set and Test set

```python
X = df["Hours"].values.reshape(-1,1)
Y = df["Scores"].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

### Fitting Simple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

### Predicting the Test set results

```python
y_pred = regressor.predict(x_test)
y_pred
```

### Visualizing Test Data

```python
plt.scatter(x_test,y_test , color = "#75a6eb")
plt.plot(x_test,y_pred , color = "black")
plt.show()
```

###  Visualization of Actual Vs Predicted Scores

```python
df1= pd.DataFrame({"Hours": x_test.reshape(1,-1)[0] , "Actual Score" : y_test.reshape(1,-1)[0] , "Predicted Score" : y_pred.reshape(1,-1)[0]})
df1
```

```python
title = "Actual Values Vs Predicted Values"
ax1 = sns.distplot(df1["Actual Score"], hist = False , color = "red" , label = "Actual Score")
sns.distplot(df1["Predicted Score"] , hist = False , color = "blue" , label = "Predicted Score" , ax = ax1)
plt.legend()
plt.grid()
plt.title(title)
plt.show()
```

### Prediction-- What will be predicted score if a student studies for 9.25 hrs/ day?

```python
hours=[[9.25]]
result= regressor.predict(hours)
print("The predicted score if a student studies for 9.25 hrs/ day is", result )
```
